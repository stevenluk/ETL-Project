```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

# Step1 Extract

```python
#extract raw data from kaggle.com
area_df=pd.read_csv("Raw Data/Area.csv")
```

```python
#extract raw data from kaggle.com
csv_file = "Raw Data/nyc-property-sales/nyc-rolling-sales.csv"
housing_data_df = pd.read_csv(csv_file)
housing_data_df.head()
```

# Step2 Transform

```python
housing_data_df['ID']=housing_data_df['BOROUGH']
```

```python
#Choose needed columns
new_housing_data_df=housing_data_df[['ID','BOROUGH','NEIGHBORHOOD','ZIP CODE','ADDRESS','SALE PRICE','SALE DATE']].copy()
new_housing_data_df
```

```python
#replace empty value with NaN
new_housing_data_df.replace("-","NaN",inplace=True)
```

```python
#Drop empty values
new_housing_data_df.dropna(subset=["SALE PRICE"],inplace=True)
```

```python
new_housing_data_df.head()
```

```python
#replace borough index into borough name
new=new_housing_data_df.replace({'BOROUGH':{1:"Manhatten",2:"Bronx",3:"Brooklyn",4:"Queens",5:"Staten Island"}})
```

```python
#drop empty value
new2=new.replace({'SALE PRICE':{" -  ":np.nan}})
```

```python
new2=new2.dropna(how='any')
```

```python
final=new2.rename(columns={'BOROUGH':'Borough','NEIGHBORHOOD':'Neighborhood','ZIP CODE':'Zip_Code','ADDRESS':'Address','SALE PRICE':'Sale_Price','SALE DATE':'Sale_Date'})
```

```python
#final=final.set_index("ID")
```

```python
final
```

```python
#add the case ID
final['Case_ID']=np.arange(1,69988)
final=final[final['Sale_Price'] > '1000']
```

```python
#read raw data
crime_df = pd.read_csv( "Raw Data/crimes-new-york-city/NYPD_Complaint_Data_Historic.csv")
crime_df.head()
```

```python
new_crime_df=crime_df[['OFNS_DESC','BORO_NM','Latitude','Longitude']].copy()
new_crime_df
```

```python
new1=new_crime_df.rename(columns={'OFNS_DESC':'Crime_Type','BORO_NM':'Borough','Latitude':'Lat','Longitude':'Lon'})
```

```python
new1['ID']=new1['Borough']
new2=new1.replace({'ID':{"MANHATTAN":1,"BRONX":2,"BROOKLYN":3,"QUEENS":4,"STATEN ISLAND":5}})
new2=new2.fillna("NaN")
new2=new2.dropna(how='any')
new2['Case_ID']=np.arange(1,1048576)
```

# Step3 Load

```python
#Check connection
connection_string = "postgres:postgres@localhost:5432/NYC Housing"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#upload data into database
area_df.to_sql(name='Area', con=engine, if_exists='append', index=False)
```

```python
#upload data into database
final.to_sql(name='Housing', con=engine, if_exists='append', index=False)
```

```python
#upload data into database
new2.to_sql(name='Crime', con=engine, if_exists='append', index=False)
```

```python
query_str = open('Query.sql')
query_text = ""
for text in query_str:
    query_text = query_text + text
      
#print(query_text)
df_query = pd.read_sql_query(query_text, con=engine)

#df_query
```
